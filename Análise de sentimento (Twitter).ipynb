```python
vacina_keyword = ('#vacina OR #vacinaja OR #vacinabrasil OR #vacinanao OR #vacinaNao OR #coronavac OR #vacinaoford')
bolsonaro_keyword = ('bolsonaro OR #bolsonaro2020 OR #forabolsonaro')
lula_keyword = ('lula OR #pt OR #forapt OR #lulalivre OR #lulapresidente')
bbb21_keyword = ('#BBB21')
```

```python
items = 2500
```

```python
#tweets_vacina = api.search(q= vacina_keyword, result_type= 'mixed')
```

```python
tweets_vacina = tw.Cursor(api.search, 
                              q= vacina_keyword,
                              since="2020-10-01", 
                              until="2021-01-25",
                              lang = 'pt'
                              ).items(items)
```

```python
tweets_bolsonaro = tw.Cursor(api.search, 
                              q= bolsonaro_keyword,
                              since="2020-10-01", 
                              until="2021-01-25",
                              lang = 'pt'
                              ).items(items)
```

```python
tweets_lula = tw.Cursor(api.search, 
                              q= lula_keyword,
                              since="2020-10-01", 
                              until="2021-01-25",
                              lang = 'pt'
                              ).items(items)
```

```python
tweets_bbb21 = tw.Cursor(api.search, 
                              q= bbb21_keyword,
                              since="2020-10-01", 
                              until="2021-01-25",
                              lang = 'pt'
                              ).items(items)
```

```python
analysis_vacina = None
analysis_bolsonaro = None
analysis_lula = None
analysis_bbb21 = None
```

```python
vacina_score = []
bolsonaro_score = []
lula_score = []
bbb21_score = []
```

```python
for tweet in tweets_vacina:
    #print('**', tweet.text)
    analysis_vacina = tb(tweet.text)
    polarity_vacina = analysis_vacina.polarity
    vacina_score.append(polarity_vacina)
```

```python
for tweet in tweets_bolsonaro:
    #print('**', tweet.text)
    analysis_bolsonaro = tb(tweet.text)
    polarity_bolsonaro = analysis_bolsonaro.polarity
    bolsonaro_score.append(polarity_bolsonaro)
```

```python
for tweet in tweets_lula:
    #print('**', tweet.text)
    analysis_lula = tb(tweet.text)
    polarity_lula = analysis_lula.polarity
    lula_score.append(polarity_lula)
```

```python
for tweet in tweets_bbb21:
    #print('**', tweet.text)
    analysis_bbb21 = tb(tweet.text)
    polarity_bbb21 = analysis_bbb21.polarity
    bbb21_score.append(polarity_bbb21)
```
